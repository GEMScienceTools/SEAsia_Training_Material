# Calculating Earthquake Recurrence for an Area Source Model - Example from Southeast Asia

The objective of this workbook is to run through a common workflow by which a catalogue is combined with a set of area source geometries to produce a PSHA input file.

In this workflow we will

i. Import and visualise some pre-defined source geometries
ii. Import and visualise a catalogue
iii. Decluster the Catalogue
iv. Analyse the Completeness of the Catalogue
v. Select the catalogue within each zone and then:
    a. Inspect hypocentral depth distribution
    b. Calculate Gutenberg-Richter recurrence parameters
vi. Write the source model to a new NRML (XML) file

```python
import numpy as np

# HMTK Parser Tools

from hmtk.parsers.catalogue.csv_catalogue_parser import CsvCatalogueParser
from hmtk.parsers.source_model.nrml04_parser import nrmlSourceModelParser

# HMTK Plotting Tools
from hmtk.plotting.mapping import HMTKBaseMap
from hmtk.plotting.seismicity.catalogue_plots import (plot_depth_histogram,
                                                      plot_magnitude_time_scatter,
                                                      plot_magnitude_time_density,
                                                      plot_magnitude_depth_density,
                                                      plot_observed_recurrence)

# HMTK utilities
from hmtk.seismicity.selector import CatalogueSelector
```

### Load and Visualise the Area Source model

```python
# Import an Area Source Model
area_source_file = '../../SEAsia_Demos/Polygons/SEAsia_simple_polygons.xml'
parser = nrmlSourceModelParser(area_source_file)
area_model = parser.read_file()

map_config = {'min_lon': 90.0, 'max_lon': 165.0, 'min_lat': -15.0, 'max_lat': 20.0, 'resolution':'l'}
# Create a hmtk basemap
basemap1 = HMTKBaseMap(map_config, 'Source Models')
# Add area sources
basemap1.add_source_model(area_model, area_border='b-')
```

### Load and Visualise the Earthquake Catalogue

##### Step 1: Decluster the Catalogue

```python
# Decluster catalogue
from hmtk.seismicity.declusterer.dec_gardner_knopoff import GardnerKnopoffType1
from hmtk.seismicity.declusterer.distance_time_windows import GardnerKnopoffWindow
# Set up the declustering algorithm
# Step 1 - set-up the tool
gardner_knopoff = GardnerKnopoffType1()
declust_config = {'time_distance_window': GardnerKnopoffWindow(),
                  'fs_time_prop': 1.0}

print 'Running declustering ...'
vcl, flag_vector = gardner_knopoff.decluster(catalogue, declust_config)
print 'done!'
print '%s clusters found' % np.max(vcl)
print '%s Non-poissionian events identified' % np.sum(flag_vector != 0)
```

```python
catalogue.select_catalogue_events(flag_vector == 0)
print 'Purged catalogue now contains %s events' % catalogue.get_number_events()
```

##### Analyse the Completeness - Stepp (1971) Method

```python
from hmtk.seismicity.completeness.comp_stepp_1971 import Stepp1971
completeness = Stepp1971()

# Create a configuration file
completeness_config = {"magnitude_bin": 0.5, "time_bin": 1.0, "increment_lock": True}

# Run Stepp method
completeness_table = completeness.completeness(catalogue, completeness_config)
for row in completeness_table:
    print row

from hmtk.plotting.seismicity.completeness.plot_stepp_1972 import create_stepp_plot
create_stepp_plot(completeness, "ExampleSteppPlot.png")
```

```python
# Look at Earthquakes within a particular zone
selector = CatalogueSelector(catalogue, create_copy=True)
for source in area_model.sources:
    # Selects the earthquakes within each polygon
    print 'Source Number %s - %s contains %8.0f events' %(source.id, source.name, source.catalogue.get_number_events())
    source.lower_depth = 75.
    source.select_catalogue(selector)
    plt.figure(figsize=(5,5))
    plot_depth_histogram(source.catalogue, 5., normalisation=True)
    print source.catalogue.get_depth_distribution(np.array([0., 25., 50., 75]), normalisation=True)
    #plt.figure(figsize=(5,5))
    #plot_observed_recurrence(source.catalogue, completeness_table, 0.1, catalogue.end_year)
    
    
```

### Look at the Earthquake Recurrence Model in Each Zone

```python
# Import the recurrence tools
from hmtk.seismicity.occurrence.utils import recurrence_table
from hmtk.seismicity.occurrence.weichert import Weichert

# Configure the recurrence module
recurrence_config = {"magnitude_interval": 0.1}

# Import the Maximum Magnitude Tool
from hmtk.seismicity.max_magnitude.cumulative_moment_release import CumulativeMoment
cumulative_moment_config = {"number_bootstraps": 100}

for source in area_model:
    print source.name
    rec_table = recurrence_table(source.catalogue.data["magnitude"], 0.1, source.catalogue.data["year"],
                                 source.catalogue.end_year - source.catalogue.start_year + 1)
    plt.figure(figsize=(5,5))
    plt.title("%s" % source.name)
    plt.bar(rec_table[:, 0], rec_table[:, 2], 0.098, color='b', log=True)
    plt.bar(rec_table[:, 0], rec_table[:, 1], 0.098, color='r', log=True)
    recurrence = Weichert()
    bval, sigmab, aval, sigma_a = recurrence.calculate(source.catalogue,recurrence_config, completeness=completeness_table)
    
    # Update Source MFD Value
    source.mfd.a_val = aval
    source.mfd.b_val = bval
    source.mfd.min_mag = 4.5
    cumulative_moment = CumulativeMoment()
    cum_mo_max_mag, _ = cumulative_moment.get_mmax(source.catalogue, mmax_config)
    print "aval = %10.4f +/- %10.4f,  b-value = %10.4f +/- %10.4f" %(aval, sigma_a, bval, sigmab)
    print "Mmax: Cumulative Moment = %8.2f" % cum_mo_max_mag
    source.mfd.max_mag = cum_mo_max_mag + 0.3
    
```
